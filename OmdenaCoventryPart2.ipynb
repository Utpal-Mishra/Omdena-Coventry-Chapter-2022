## **Visualization**

```
ethnicity = data.melt(id_vars=["Ethnicity"], var_name = "Year", value_name = "Percentage")
# ethnicity.sort_values(["Year"], inplace = True)
ethnicity
```

```
fig = px.bar(ethnicity, x="Ethnicity", y="Percentage", color="Ethnicity", barmode="group", 
            # facet_row="Mental Status", 
             facet_col="Year")
fig.update_xaxes(rangeslider_visible=True, showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_yaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
# fig.update_traces(marker_color='rgb(158,202,225)', marker_line_color='rgb(8,48,107)', marker_line_width=1.5, opacity=0.6)
fig.update_layout(height=600, width=1560, title_text="Percentage of Diverse Ethnicity in The UK")
fig.show()
```
